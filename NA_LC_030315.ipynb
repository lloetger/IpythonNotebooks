```python
# import time module to measure execution time

import time
tic = time.time() 

# import fsl
import nipype.interfaces.fsl as fsl

# select subject
brain_no=1

# print working directory
import os
pwdirectory='/mnt/dzne/loetgeringl/NA_LC/00'+str(brain_no)
os.chdir(pwdirectory)
# utils directory
utilsdirectory='/mnt/dzne/loetgeringl/LC_data/Utils'
print pwdirectory
```

### 1) Reorient image to match approximate MNI152 orientation

```python
# MPRAGE
reorient=fsl.Reorient2Std()
reorient.inputs.in_file = "MPRAGE.nii"
reorient.inputs.output_type="NIFTI"
res = reorient.run()
```

```python
# robustfov is not a module of the fsl interface in nipype
# therefore execute robustfov in unix shell
os.popen('robustfov -i MPRAGE_reoriented -r MPRAGE_reoriented_roi').read()
```

### 3) Bias field correction (FSL FAST)

```python
# raplace NaN's with zeros
os.popen('fslmaths MPRAGE_reoriented_roi -nan MPRAGE_reoriented_roi_noNAN').read()

os.popen('N4BiasFieldCorrection -i MPRAGE_reoriented_roi_noNAN.nii.gz -o MPRAGE_reoriented_roi_restore.nii.gz').read()


# measure elapsed time up to this point
toc = time.time() - tic
print toc
```

### 4) Brain extraction (FSL BET)

```python
bet=fsl.BET()
bet.inputs.in_file = 'MPRAGE_reoriented_roi_restore.nii.gz'
bet.inputs.out_file = 'MPRAGE_reoriented_roi_restore_bet.nii.gz'
# fractional intensity threshold (0->1)
bet.inputs.frac=0.2
result = bet.run()
```

### 5) Affine MPRAGE Coregistration (FLIRT)

```python
# - register MPRAGE onto MNI brain
# - calculate matrix to transform from MNI space to MPRAGE space
# - transform Keren atlas into MPRAGE space

flt = fsl.FLIRT(bins=256, cost_func='corratio')
flt.inputs.in_file = 'MPRAGE_reoriented_roi_restore_bet.nii.gz'
flt.inputs.reference = utilsdirectory+'/MNI152_T1_1mm_brain.nii.gz'
flt.inputs.out_file = 'MPRAGE_in_MNI.nii.gz'
flt.inputs.out_matrix_file = 'MP2MNI.mat'
flt.inputs.dof=12
flt.run()
os.popen('convert_xfm -omat MNI2MP.mat -inverse MP2MNI.mat');
```

### 6) Nonlinear Coregistration (FNIRT)

#### Calculate warp field

```python
fnt=fsl.FNIRT()
fnt.inputs.in_file='MPRAGE_reoriented_roi_restore_bet.nii.gz'
fnt.inputs.ref_file=utilsdirectory+'/MNI152_T1_1mm_brain.nii.gz'
fnt.inputs.affine_file='MP2MNI.mat'
fnt.inputs.regularization_model='bending_energy'
fnt.inputs.regularization_lambda=[300,100,50,30]
fnt.inputs.subsampling_scheme=[16,8,4,2]
fnt.run()

# OUTPUT: MPRAGE_bet_restore_warpcoef.nii.gz
# OUTPUT: MPRAGE_bet_restore_warped.nii.gz
```

#### Invert warp field

```python
iwarp=fsl.InvWarp()
iwarp.inputs.reference='MPRAGE_reoriented_roi_restore_bet.nii.gz'
iwarp.inputs.warp='MPRAGE_reoriented_roi_restore_bet_warpcoef.nii.gz'
iwarp.run()
```

#### Apply inverted warp field to Keren Atlas (2SD)

```python
awarp=fsl.ApplyWarp()
awarp.inputs.ref_file='MPRAGE_reoriented_roi_restore_bet.nii.gz'
awarp.inputs.in_file=utilsdirectory+'/LC_2SD_BINARY_TEMPLATE_1mm.nii.gz'
awarp.inputs.field_file='MPRAGE_reoriented_roi_restore_bet_warpcoef_inverse.nii.gz'
awarp.inputs.out_file='LC_2SD_BINARY_TEMPLATE_1mm_MPRAGE.nii.gz'
awarp.run()
```

#### Apply inverted warp field to Keren Atlas (2SD) (left and right separately)

```python
# left

awarp=fsl.ApplyWarp()
awarp.inputs.ref_file='MPRAGE_reoriented_roi_restore_bet.nii.gz'
awarp.inputs.in_file=utilsdirectory+'/LC_2SD_BINARY_TEMPLATE_1mm_left.nii.gz'
awarp.inputs.field_file='MPRAGE_reoriented_roi_restore_bet_warpcoef_inverse.nii.gz'
awarp.inputs.out_file='LC_2SD_BINARY_TEMPLATE_1mm_MPRAGE_left.nii.gz'
awarp.run()

# right

awarp=fsl.ApplyWarp()
awarp.inputs.ref_file='MPRAGE_reoriented_roi_restore_bet.nii.gz'
awarp.inputs.in_file=utilsdirectory+'/LC_2SD_BINARY_TEMPLATE_1mm_right.nii.gz'
awarp.inputs.field_file='MPRAGE_reoriented_roi_restore_bet_warpcoef_inverse.nii.gz'
awarp.inputs.out_file='LC_2SD_BINARY_TEMPLATE_1mm_MPRAGE_right.nii.gz'
awarp.run()
```

```python
%matplotlib inline
import math
import numpy as np
from lmfit import minimize, Parameters

# define residual function

def residual(params, I, data):
    amp_tissue = params['amp_tissue'].value
    int_tissue = params['int_tissue'].value
    sigma_tissue = params['sigma_tissue'].value
    
    amp_lc = params['amp_lc'].value
    int_lc = params['int_lc'].value
    sigma_lc = params['sigma_lc'].value
    
    model = amp_tissue*np.exp(-(I-int_tissue)**2/(2*sigma_tissue**2))+amp_lc*np.exp(-(I-int_lc)**2/(2*sigma_lc**2))

    return (data-model)

print 'min='+str(np.min(data)) 
print 'max='+str(np.max(data))
print data.shape
data2, bins=np.histogram(data,bins=np.max(data)-np.min(data))

# initial parameters

params = Parameters()
params.add('amp_tissue', value=10,min=1)
params.add('int_tissue', value=mean(data),max=np.max(data))
params.add('sigma_tissue', value=std(data), min=0)
params.add('amp_lc', value=1,min=1)
params.add('int_lc', value=mean(data)+std(data),min=(params['int_tissue']),max=np.max(data))
params.add('sigma_lc', value=std(data), min=0)


# find paramters to minimize linear model


out = minimize(residual, params,'leastsq', args=(bins[0:(max(data)-min(data))], data2))
```
